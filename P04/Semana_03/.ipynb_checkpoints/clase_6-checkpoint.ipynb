# Ejemplo de una clase principal (padre, main)

```python
# ejemplo de clase main en POO clase medios de transporte
# profe jose
# 01/02/24
# ejemplo de como utilizar clases en python

#_________________________________clase principal___________________________

class medios_de_transporte: # clase main , clase abuelo
    
    # SIEMPRE!! comienzo con el constructor 
    # en el constructor asigno los atributos del objeto que pertenece a la clase
    
    def __init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color): #constructor
        #self.atributo=atributo
        self.capacidad_pasajeros=capacidad_pasajeros
        self.vel_max=vel_max
        self.combustible=combustible
        self.peso=peso
        self.marca=marca
        self.material=material
        self.color=color
       
    #_____________________ printer_________________# imprimir bonito
    
    def __str__(self):
        return f"Atributos del objeto:\n capacidad_pasajeros:{self.capacidad_pasajeros}\n vel_max:{self.vel_max}\n combustible:{self.combustible} \n peso:{self.peso} \n marca:{self.marca} \n material:{self.material} \n color:{self.color}"
       
    
    #______________________getters _________________
    #get -> obtener -> obter los atributos del objeto
   
    def get_capacidad_pasajeros(self):
        return self.capacidad_pasajeros
    def get_vel_max(self):
        return self.vel_max
    def get_combustible(self):
        return self.combustible
    def get_peso(self):
        return self.peso
    def get_marca(self):
        return self.marca
    def get_material(self):
        return self.material
    def get_color(self):
        return self.color
    
    #______________________setters _________________
    #set -> configurar -> modificar los atributos del objeto
    
    def set_capacidad_pasajeros(self,capacidad_pasajeros):
         self.capacidad_pasajeros=capacidad_pasajeros
    def set_vel_max(self,vel_max):
         self.vel_max=vel_max
    def set_combustible(self,combustible):
         self.combustible=combustible        
    def set_peso(self,peso):
        self.peso=peso
    def set_marca(self,marca):
        self.marca=marca
    def set_material(self,material):
        self.material=material
    def set_color(self,color):
        self.color=color
   
 #_____________________ pruebas del codigo  ____________________________  

#para crear un objeto ponemos el nombre de la clase (atributos)
#capacidad_pasajeros,vel_max,combustible,peso,marca,material,color
carro=medios_de_transporte(5,180,"gasolina",500,"toyota","acero","rojo")
lancha=medios_de_transporte(10,240,"diesel",500,"sdfghjk","fibra_de_vidrio","blanco")
print(carro)
print("______________________")
print(lancha)
print("______________________")

list_objetos=[carro,lancha]

for objeto in list_objetos:
    print(objeto.get_color())
    
carro.set_color("verde")    
    
print("______________________")

list_objetos=[carro,lancha]

for objeto in list_objetos:
    print(objeto.get_color())
       

```

```python
#_________________________________clase principal___________________________

class medios_de_transporte: # clase main , clase abuelo
    
    # SIEMPRE!! comienzo con el constructor 
    # en el constructor asigno los atributos del objeto que pertenece a la clase
    
    def __init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color): #constructor
        #self.atributo=atributo
        self.capacidad_pasajeros=capacidad_pasajeros
        self.vel_max=vel_max
        self.combustible=combustible
        self.peso=peso
        self.marca=marca
        self.material=material
        self.color=color
       
    #_____________________ printer_________________# imprimir bonito
    
    def __str__(self):
        return f"Atributos del objeto:\n capacidad_pasajeros:{self.capacidad_pasajeros}\n vel_max:{self.vel_max}\n combustible:{self.combustible} \n peso:{self.peso} \n marca:{self.marca} \n material:{self.material} \n color:{self.color}"
       
    
    #______________________getters _________________
    #get -> obtener -> obter los atributos del objeto
   
    def get_capacidad_pasajeros(self):
        return self.capacidad_pasajeros
    def get_vel_max(self):
        return self.vel_max
    def get_combustible(self):
        return self.combustible
    def get_peso(self):
        return self.peso
    def get_marca(self):
        return self.marca
    def get_material(self):
        return self.material
    def get_color(self):
        return self.color
    
    #______________________setters _________________
    #set -> configurar -> modificar los atributos del objeto
    
    def set_capacidad_pasajeros(self,capacidad_pasajeros):
         self.capacidad_pasajeros=capacidad_pasajeros
    def set_vel_max(self,vel_max):
         self.vel_max=vel_max
    def set_combustible(self,combustible):
         self.combustible=combustible        
    def set_peso(self,peso):
        self.peso=peso
    def set_marca(self,marca):
        self.marca=marca
    def set_material(self,material):
        self.material=material
    def set_color(self,color):
        self.color=color
   
 #_________________________SUBCLASES_______________

class transporte_terrestre(medios_de_transporte):
    #_____________constructor______________
    #            self, atributos heredados, atributos nuevos
    def __init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color,n_ruedas,retrovisores,faros):
            #heredo el constructor de la clase main
            medios_de_transporte.__init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color)
            #agrego los nuevos atributos
            self.n_ruedas=n_ruedas
            self.retrovisores=retrovisores
            self.faros=faros
    #_____________printer__________________
    def __str__(self):
        return f"Atributos del objeto:\n capacidad_pasajeros:{self.capacidad_pasajeros}\n vel_max:{self.vel_max}\n combustible:{self.combustible} \n peso:{self.peso} \n marca:{self.marca} \n material:{self.material} \n color:{self.color} \n n_ruedas:{self.n_ruedas} \n retrovisores:{self.retrovisores} \n faros:{self.faros}"
     
    #_____________getters__________________
    def get_n_ruedas(self):
        return self.n_ruedas
    def get_retrovisores(self):
        return self.retrovisores
    def get_faros(self):
        return self.faros
    #_____________setters__________________
    def set_n_ruedas(self,n_ruedas):
        self.n_ruedas=n_ruedas
    def set_retrovisores(self,retrovisores):
        self.retrovisores=retrovisores
    def set_faros(self,faros):
        self.faros=faros
    
class transporte_aereo(medios_de_transporte):
    #_____________constructor______________
    def __init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color,n_rotores,n_alerones,altitud_max):
            #heredo el constructor de la clase main
            medios_de_transporte.__init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color)
            #agrego los nuevos atributos
            self.n_rotores=n_rotores
            self.n_alerones=n_alerones
            self.altitud_max=altitud_max
    #_____________printer__________________
    def __str__(self):
        return f"Atributos del objeto:\n capacidad_pasajeros:{self.capacidad_pasajeros}\n vel_max:{self.vel_max}\n combustible:{self.combustible} \n peso:{self.peso} \n marca:{self.marca} \n material:{self.material} \n color:{self.color} \n n_rotores:{self.n_rotores} \n n_alerones:{self.n_alerones} \n altitud_max:{self.altitud_max}"
       
    #_____________getters__________________
    def get_n_rotores(self):
        return self.n_rotores
    def get_n_alerones(self):
        return self.n_alerones
    def get_altitud_max(self):
        return self.altitud_max
    #_____________setters__________________
    def set_n_rotores(self,n_rotores):
        self.n_rotores=n_rotores
    def set_n_alerones(self,n_alerones):
        self.n_alerones=n_alerones
    def set_altitud_max(self,altitud_max):
        self.altitud_max=altitud_max
    
class transporte_acuatico(medios_de_transporte):
    #_____________constructor______________
    def __init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color,n_rotores,n_anclas):
            #heredo el constructor de la clase main
            medios_de_transporte.__init__(self,capacidad_pasajeros,vel_max,combustible,peso,marca,material,color)
            #agrego los nuevos atributos
            self.n_rotores=n_rotores
            self.n_anclas=n_anclas
    #_____________printer__________________
    def __str__(self):
        return f"Atributos del objeto:\n capacidad_pasajeros:{self.capacidad_pasajeros}\n vel_max:{self.vel_max}\n combustible:{self.combustible} \n peso:{self.peso} \n marca:{self.marca} \n material:{self.material} \n color:{self.color} \n n_rotores:{self.n_rotores} \n n_anclas:{self.n_anclas}"
       
    #_____________getters__________________
    def get_n_rotores(self):
        return self.n_rotores
    def get_n_anclas(self):
        return self.n_anclas
    #_____________setters__________________

    def set_n_rotores(self,n_rotores):
        self.n_rotores=n_rotores
    def set_n_anclas(self,n_anclas):
        self.n_anclas=n_anclas

 #_____________________ pruebas del codigo  ____________________________

# terrestre:capacidad_pasajeros,vel_max,combustible,peso,marca,material,color,n_ruedas,retrovisores,faros):
moto=transporte_terrestre(2,160,"gasolina",150,"honda","acero","negro",2,2,1)
# aereo:capacidad_pasajeros,vel_max,combustible,peso,marca,material,color,n_rotores,n_alerones,altitud_max):
helicoptero=transporte_aereo(6,100,"diesel",800,"asd","acero","negro",2,4,4000)        
# acuatico:capacidad_pasajeros,vel_max,combustible,peso,marca,material,color,n_rotores,n_anclas):
crucero=transporte_acuatico(1000,70,"diesel",900000,"naval","acero","blanco",3,4) 

print(moto)
print("_____________")
print(helicoptero)
print("_____________")
print(crucero)
print("_____________")
print()

lista_medios=[moto,helicoptero,crucero]

for objeto in lista_medios:
   
    print(f"vel max :{objeto.get_vel_max()}")
       
```
