```python
# me piden un programa que devuelva el area del circulo varias veces variando el radio

#necesito que me diga el area de el circulo para un radio r
# para un radio 2r,3r,4r,5r...... 100r

# //////////////////////////////////////////
# Calculadora de Area del circulo
# 30/01/24
# profe Jose
# proceso para hallar el area de un circulo segun n veces su radio
# //////////////////////////////////////////
# _______funciones_________

def area(r):
    result=3.14*(r**2)
    return result

#_________ Main _____________

radio=int(input("ingrese el radio del circulo"))

#resultado=area(radio)

multiplos=list(range(1,101,1))

for n in multiplos:
    resultado=area(radio*n)
    print(f"el area  con r={radio*n} es A={resultado}")
```

ejemplo

* subclases : una clase que pertenece a una clase mayor
ejemplo

* OBJETOS: (instancias), es un elemento que pertenece a una clase y poseen unos atributos
