# Abrir y usar el modelo

```python
import pickle
f=open("modelo_final.pickle","rb")
load_model=pickle.load(f)
```

```python
# el sigueinte paso seria ingresar la data que se queire pedecir

from IPython.core.display import HTML
from sklearn.datasets import load_iris

# cargo la "nueva" data a la que aplicar√© ML para clasificar
datos=load_iris()
datos.data
```

```python
# hago la prediccion

predicted_species=load_model.predict(datos.data)
```

```python
import pandas as pd

#datos["data"]
#datos.data
# feature_names -> nombres de las columnas de la data
data_df=pd.DataFrame(data=datos.data,columns=datos.feature_names)
#display(HTML(data_df.to_html()))

target=pd.DataFrame(data=predicted_species,columns=["species"])
#display(HTML(target.to_html()))

iris=pd.concat([data_df,target],axis=1)
display(HTML(iris.to_html()))
```

```python
#Guardando los dataframes en un archivo .csv

from datetime import datetime

date = datetime.now().strftime("%d-%m-%Y")

def guardarCSV(df, nombre, fecha):
    nombre = nombre + "_" + fecha + '.csv'
    print(nombre)
    df.to_csv(nombre, index = True)
    
guardarCSV(iris,"Predicciones_Iris",date)   
    
```
