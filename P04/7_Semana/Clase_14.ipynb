```python
# import pandas as pd 
from IPython.core.display import HTML
#display(HTML(df_data.to_html())) # print
lista1=[1,2,3,4,5]
lista2=[6,7,8,9,10]
s_l1=pd.Series(lista1)
s_l2=pd.Series(lista2)
data={"col1":s_l1,"col2":s_l2}
df_data=pd.DataFrame(data)
display(HTML(df_data.to_html())) # print

############################ utilidades para un dataframe #############


####_____________________ asignar nombres a filas y columnas _________

#index para las filas  index=0
df_data.index=["chocolates","bombones","gomitas","papitas","refresco"]
display(HTML(df_data.to_html())) # print

#index para las columnas  index=1 -> columns
df_data.columns=["estante1","estante2"]
display(HTML(df_data.to_html())) # print

####_____________________ reemplazar nombres a filas y columnas _________

#cuando queiro que los cambios se vean reflejados en el df original
#de manera inmediata, debo poner inplace=True
df_data.rename(index={"bombones":"caramelos"},inplace=True)
df_data.rename(columns={"estante2":"bodega"},inplace=True)
display(HTML(df_data.to_html())) # print


####_____________________ modificar datos _________
print("___________modificar dato ______________")
df_data.loc["caramelos","estante1"]= 10
display(HTML(df_data.to_html())) # print
print("___________ otro modo de modificar dato ______________")
df_data.iloc[1,0]= 20
display(HTML(df_data.to_html())) # print

####_____________________ acceder a la informacion _________
print(len(df_data)) # tamaño de las filas
print(df_data.shape) # tamaño: filas x columnas
print(df_data.index) 
print(df_data.columns) 
print(df_data.values) 
print(type(df_data.values))
```

```python
import pandas as pd 
from IPython.core.display import HTML
#display(HTML(df_data.to_html())) # print

df = pd.DataFrame({
    "fecha":pd.date_range(start='2024-03-07',periods=9),
    "pais":["colombia","panama","costa rica"]*3,
    "ventas":range(1,10)      
})
print("__________DF original stackeado_________")
display(HTML(df.to_html())) # print

# unstack el dataframe : desapilar
#                                 0            1
# la sintaxis es con comilla ' 
df_unstacked = df.set_index(['fecha', 'pais'])['ventas'].unstack('pais')
# el unstack se setea como colummna
#index seria fecha y pais
# values las ventas


print("__________DF desapilado_________")
display(HTML(df_unstacked.to_html())) # print
```

```python
import pandas as pd 
from IPython.core.display import HTML
#display(HTML(df_data.to_html())) # print

#importar la data con la que vamos a trabajar

file="train.csv"

def open_csv(file):
    data=pd.read_csv(file) #SIEMPRE como un df
    return data

df=open_csv(file)
display(HTML(df.to_html())) # print
```

```python
#accedo a una sola fila como serie

df2=df.iloc[2] # ,cuando saco 1 sola fila ya no es un df , es una serie
print(type(df2))
df2
```

```python
#accedo a una sola columna como serie

edades=df["Age"]

avg_edades=edades.mean()
print("promedio segun .mean():",avg_edades)
print("suma de las edades:", edades.sum())
print("suma / 714 :",edades.sum()/714)
print("suma / 891 :",edades.sum()/891)
# segun .info() edades tiene 714 de 891 datos
print("conclusion: pandas descarta los valores nulos \n cuando hace el analisis estadistico con .describe()")
```

```python
#df esta el df original
#en df_new, los cambios realizados en la explicacion de clase

import matplotlib.pyplot as plt
edades=df["Age"] # edades al ser sacado como columna es una serie
plt.bar(edades.index,edades.values)
plt.show()
```

```python
# otra manera de llamar a las graficas
#otra manera estandar de llamar a las graficas
df.Age.plot(kind="bar",title="Grafico de barras de edades")
plt.show()
```

```python
# Histograma de edades
#otra manera estandar de llamar a las graficas
df.Age.plot(kind="hist",title="Grafico de barras de edades")
plt.show()
```

```python
# grafico circular de sobrevivientes
#otra manera estandar de llamar a las graficas

# Creamos la figura y los ejes
fig, ax = plt.subplots()
# Diagrama de sectores de falleccidos y supervivientes
#value_counts, saca el dato, y cuantas veces se repite
df.Survived.value_counts().plot(kind = "pie", labels = ["Fallecidos", "Supervivientes"], title = "Distribución de supervivientes")
plt.show()
```

```python
# Diagrama de barras con el número de personas de cada clase
df.groupby("Pclass").size().plot(kind = "bar", title = "Número de personas por clase")
plt.show()
```

```python
# Diagrama de barras con el número de personas fallecidas y supervivientes de cada clase
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", title = "Número de personas fallecidas y supervivientes por clase")
plt.show()
```

```python
# Diagrama de barras con el número de personas fallecidas y supervivientes acumuladas de cada clase
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", stacked = True, title = "Número de personas fallecidas y supervivientes por clase")
plt.show()
```
