# TensorFlow, Keras y tf.keras

### TF
TensorFlow es una librería de código abierto desarrollada por Google que se utiliza para realizar cálculos numéricos utilizando gráficos de flujo de datos. Se utiliza principalmente para tareas de aprendizaje automático y deep learning.

TensorFlow se utiliza para una variedad de aplicaciones en aprendizaje automático, como clasificación de imágenes, reconocimiento de voz, procesamiento de texto, series temporales, entre otros.

### keras

Keras es una API de aprendizaje profundo (una herramienta que permite que dos aplicaciones intercambien datos) escrita en Python.
Se ejecuta sobre TensorFlow.
Fue desarrollado con un enfoque en permitir la experimentación rápida

### TF.Keras

tf.keras es una implementación de Keras específica de Tensorflow. El prefijo del nombre "tf" dice que es específico del marco TensorFlow.


### Tensores

Los tensores son las estructuras de datos primarias utilizadas por las redes neuronales complejas

En las redes neuronales, las transformaciones, entrada, salida, etc., se realizan mediante tensores.

¿Qué son los tensores?

Los tensores son arreglos multidimensionales con un tipo uniforme (llamado tipo), los tensores son (más o menos) como matrices NumPy.

Osea que, los tensores son una representación matematica estructurada de una red neuronal que contiene todos los elementos vistos en la teoria basica de una neurona
Pueden exisitr tensores que contengan una red n-dimensional de redes neuronales

Todos los tensores son inmutables como los números y las cadenas de Python, es decir, nunca puede actualizar el contenido de un tensor, solo puede crear uno nuevo.

# Ejemplo 1 
c° a F°

1 Tensor: 
Red 1 capa, 1 neurona

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

temperature_df = pd.read_csv("celsius_a_fahrenheit.csv")
sns.scatterplot(temperature_df['Celsius'], temperature_df['Fahrenheit'])

```

```python
# Cargando Set de Datos
X_train = temperature_df['Celsius'] 
y_train = temperature_df['Fahrenheit'] 

# Creando el Modelo
# https://keras.io/guides/sequential_model/
model = tf.keras.Sequential() # Se crea un modelo secuencial de Keras.

# Se agrega una capa completamente conectada (dense) a nuestro modelo.
#La capa tiene un solo nodo (units=1) (una sola neurona en una capa) y
# espera un input con forma [1] (un solo valor de entrada).
model.add(tf.keras.layers.Dense(units=1, input_shape=[1])) 


model.summary() 
# Esta línea genera un resumen de la arquitectura del modelo,
#incluyendo el número de capas, el número de parámetros y la forma de
#entrada y salida de cada capa.
```

```python
#optimización
#                                          learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss='mean_squared_error') 
# Se compila el modelo, especificando el optimizador Adam con un learning rate de 1
# y la función de pérdida como mean squared error.
```

```python
#Entrenando el modelo
epochs_hist = model.fit(X_train, y_train, epochs = 100) 
# El modelo se entrena utilizando los datos de entrenamiento, con 100 épocas.

```

El número de épocas se refiere a cuántas veces el modelo se entrena en todo el conjunto de datos de entrenamiento.

Si el número de épocas es demasiado bajo, el modelo puede no tener suficiente tiempo de entrenamiento para aprender los patrones complejos en los datos, lo que resulta en un subajuste. El subajuste ocurre cuando el modelo no puede capturar los patrones subyacentes en los datos, lo que resulta en un rendimiento deficiente tanto en los datos de entrenamiento como en los de prueba.

Por otro lado, si hay demasiadas épocas, el modelo puede memorizar el conjunto de entrenamiento, lo que lleva a un sobreajuste. El sobreajuste ocurre cuando un modelo funciona mal en los datos de prueba porque es demasiado complejo y comienza a ajustarse al ruido en los datos. Esto se debe a que el modelo no puede generalizar a nuevos datos debido a su aprendizaje excesivo de los datos de entrenamiento.


```python
#Evaluando modelo
keys =epochs_hist.history.keys() # Se imprimen las claves del diccionario 
# 'history', que contiene métricas de entrenamiento y validación, 
#como la pérdida y la precisión en cada época.
keys
```

```python
#Grafico
plt.plot(epochs_hist.history['loss'])
plt.title('Progreso de Perdida durante Entrenamiento del Modelo')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend('Training Loss')


```

```python
#Predicciones
# creo una lista para probar, en base a la lista original
# original : X_train = temperature_df['Celsius'] 
x_test=X_train[::2]
# original : y_train = temperature_df['Fahrenheit'] 
y_test=y_train[::2]
print(x_test)
```

```python
#Predicciones
Temp_F = model.predict([x_test])
pred=pd.DataFrame(Temp_F)
pred.columns = ['Pred']
real=pd.DataFrame(y_test)
real = real.reset_index(drop=True)

result = pd.concat([pred, real], axis=1)
result
```

# Ejemplo 2
Prediccion de ganacia en ventas segun la temperatura (clima) para una empresa de helados

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Importando Datos
sales_df = pd.read_csv("datos_de_ventas.csv")

#Visualizacion
sns.scatterplot(sales_df['Temperature'], sales_df['Revenue'])


```

```python

#Creando set de entrenamiento
X_train = sales_df['Temperature']
y_train = sales_df['Revenue']

#Creando Modelo
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))

model.summary()

#model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error', \
              metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])


#Entrenamiento
epochs_hist = model.fit(X_train, y_train, epochs = 1000)


keys = epochs_hist.history.keys()


```

```python
# Grafico de Entrenamiento del Modelo
plt.plot(epochs_hist.history['loss'])
plt.title('Progreso de Perdida durante Entrenamiento')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
#Predicciones
# creo na lista para probar, en base a la lista original
# original : X_train = temperature_df['Celsius'] 
x_test=X_train[::2]
# original : y_train = temperature_df['Fahrenheit'] 
y_test=y_train[::2]


#Predicciones
Revenue = model.predict([x_test])
pred=pd.DataFrame(Revenue)
pred.columns = ['Pred']
real=pd.DataFrame(y_test)
real = real.reset_index(drop=True)

result = pd.concat([pred, real], axis=1)
result
```

```python
#Grafico de Prediccion
plt.scatter(x_test, y_test, color='gray')
plt.plot(x_test, model.predict(x_test), color = 'red')
plt.ylabel('Ganancia [Dolares]')
plt.xlabel('Temperatura [gCelsius]')
plt.title('Ganancia Generada vs. Temperatura @Empresa de Helados')
```

```python
#evaluar el rendimiento del modelo

loss, accuracy, precision, recall = model.evaluate(x_test, y_test)
print(f'Test loss: {loss:.3f}')
print(f'Test accuracy: {accuracy:.3f}')
print(f'Test precision: {precision:.3f}')
print(f'Test recall: {recall:.3f}')
```

En la evaluación de rendimiento de un modelo de aprendizaje profundo, los parámetros de rendimiento más comunes son:

* Loss: Es una medida de cuánto se desvía el modelo de las etiquetas verdaderas en el conjunto de pruebas. El loss se calcula como la suma de las pérdidas de cada ejemplo en el conjunto de pruebas. Cuanto más bajo sea el valor de loss, mejor será el rendimiento del modelo.

* Accuracy: Es la proporción de ejemplos en el conjunto de pruebas que el modelo clasificó correctamente. La accuracy se calcula como el número de ejemplos clasificados correctamente dividido por el número total de ejemplos en el conjunto de pruebas. Cuanto más alto sea el valor de accuracy, mejor será el rendimiento del modelo.

* Precision: Es la proporción de ejemplos positivos reales que el modelo clasificó correctamente. La precision se calcula como el número de ejemplos positivos reales clasificados correctamente dividido por el número total de ejemplos positivos predichos por el modelo. Cuanto más alto sea el valor de precision, mejor será el rendimiento del modelo en predecir positivos.

* Recall: Es la proporción de ejemplos positivos reales que el modelo clasificó correctamente en relación con el número total de ejemplos positivos reales. El recall se calcula como el número de ejemplos positivos reales clasificados correctamente dividido por el número total de ejemplos positivos reales. Cuanto más alto sea el valor de recall, mejor será el rendimiento del modelo en detectar positivos.

En resumen, loss mide el error general del modelo, la accuracy mide la proporción de ejemplos clasificados correctamente, la precision mide la proporción de ejemplos positivos predichos correctamente, y el recall mide la proporción de ejemplos positivos reales detectados por el modelo.
