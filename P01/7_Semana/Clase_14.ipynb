```python
import pandas as pd
from IPython.core.display import HTML
l1=[2,34,56,7,4]
l2=[5,3,578,4,6]
s_l1=pd.Series(l1)
s_l2=pd.Series(l2)

##################################definir  un df

data={"col1":s_l1,"col2":s_l2}
df_data=pd.DataFrame(data)
display(HTML(df_data.to_html())) # print

#################################asignar nombres a filas y columnas
#index para las filas
df_data.index=["id_galletas","id_bombom","id_chocolates","id_papitas","id_gomita"]
display(HTML(df_data.to_html())) # print
print("_________________________")
#columns para las columnas
df_data.columns=["tienda_1","tienda_2"]
display(HTML(df_data.to_html())) # print
print("_________________________")

################################ reemplazar nombres de filas y columnas

#siempre que trabajo sobre el mismo dataframe -> inplace=True para que efectue los cambios de inmediato
df_data.rename(index={"id_galletas":"numero_galletas","id_papitas":"numero_papitas"},inplace=True)
df_data.rename(columns={"tienda_1":"bodega_1"},inplace=True)
display(HTML(df_data.to_html())) # print
print("_________________________")

######################### modificar un dato
print("modificar un dato")
df_data.loc["id_chocolates", "tienda_2"] = 37
display(HTML(df_data.to_html())) # print
print("_________________________")


############################## Acceder a la informacion

print(len(df_data)) # tamaño filas
print(df_data.shape) # tamaño filas x col
print(df_data.index) #
print(df_data.columns) #
print(df_data.values) #
print(type(df_data.values)) #

```

```python
import pandas as pd
from IPython.core.display import HTML
import pandas as pd


df = pd.DataFrame({
    'fecha': pd.date_range(start='2023-01-01', periods=9),
    'pais': ['EE. UU.', 'Reino Unido', 'CA'] * 3,
    'producto': ['A', 'B', 'C'] * 3,
    'ventas': range(1, 10)
})
print("__________df original__________")
display(HTML(df.to_html())) # print

## Unstack the DataFrame
df_unstacked = df.set_index(['fecha', 'pais'])['ventas'].unstack('pais')

print("__________df desapilado__________")
display(HTML(df_unstacked.to_html())) # print

```

```python
import pandas as pd
from IPython.core.display import HTML
import matplotlib.pyplot as plt

file="train.csv"

def open_csv(file):
    data=pd.read_csv(file) # siempre lo lee como dataframe
    #print(type(data))
    return data

df=open_csv(file)
avg_age=df["Age"].mean(axis=0)
edades=df["Age"]

plt.bar(edades.index,edades.values)
plt.show
```

```python
# Histograma de edades
df.Age.plot(kind = "hist", title = "Histograma de edades")
plt.show()
```

```python
import matplotlib.pyplot as plt
# Creamos la figura y los ejes
fig, ax = plt.subplots()
# Diagrama de sectores de falleccidos y supervivientes
df.Survived.value_counts().plot(kind = "pie", labels = ["Fallecidos", "Supervivientes"], title = "Distribución de supervivientes")
plt.show()
```

```python
# Diagrama de barras con el número de personas de cada clase
df.groupby("Pclass").size().plot(kind = "bar", title = "Número de personas por clase")
plt.show()
```

```python
# Diagrama de barras con el número de personas fallecidas y supervivientes de cada clase
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", title = "Número de personas fallecidas y supervivientes por clase")
plt.show()
```

```python
# Diagrama de barras con el número de personas fallecidas y supervivientes acumuladas de cada clase
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", stacked = True, title = "Número de personas fallecidas y supervivientes por clase")
plt.show()
```
