### EJEMPLO DE UNA CLASE PRINCIPAL 

```python
# ejemplo sobre POO clase medios de transporte
# profe jose
# 31/01/24
# ejemplo de manejo de clases, constructores, autoparametros, setters, y getters


#_______________clase principal_________________________

class medios_de_transporte: # clase main(principal) , clase abuelo
    
    #siempre defino de primero el constructor
    #En el constructor asigno los atributos del objeto que pertenece a esa clase
    
    def __init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible): #constructor
        #self.atributo=atributo
        self.capacidad_pasajeros=capacidad_pasajeros
        self.color=color
        self.tamaño=tamaño
        self.potencia=potencia
        self.combustible=combustible
        
        
    # a casi todas las clases SIEMPRE le hago 3 funciones basicas
    
    #________________________PRINT________________________________
    # IMPRIME LOS ATRIBUTOS DEL OBJETO
    
    def __str__(self):
        
        return f"atributos del objeto : {self.capacidad_pasajeros},{self.color},{ self.tamaño},{self.potencia},{self.combustible}"
    
    
    
    # ______________________SETTERS_______________________________
    # set -> SETUP -> CONFIGURAR O MODIFICAR ALGUN ATRIBUTO DEL OBJETO
    
    def set_capacidad_pasajeros(self,capacidad_pasajeros):
        self.capacidad_pasajeros=capacidad_pasajeros
    def set_color(self,color):
        self.color=color
    def set_tamaño(self,tamaño):
        self.tamaño=tamaño
    def set_potencia(self,potencia):
        self.potencia=potencia
    def set_combustible(self,combustible):
        self.combustible=combustible
        
    
    #________________________GETTERS______________________________
    # get -> obtener, para obtener un valor de un atributo
    
    def get_capacidad_pasajeros(self):
        return self.capacidad_pasajeros
    def get_color(self):
        return self.color
    def get_tamaño(self):
        return self.tamaño
    def get_potencia(self):
        return self.potencia
    def get_combustible(self):
        return self.combustible


#______________________________ejecucion para ver el proceso__________________________________________        
 
 #crear un objeto, ponemos el nombre de la clase
carro1=medios_de_transporte(4,"rojo",8000,2000,"gasolina")
print(carro1)

carro1.color="verde"
print(carro1.color)
print("________________________________")

print(carro1.get_color())
carro1.set_color("negro")
print(carro1.get_color())
```

### caso 1 -> Heredo todo de la clase padre sin poner nada nuevo en las subclases : constructor , atributos, metodos

```python
#_______________clase principal_________________________

class medios_de_transporte: # clase main(principal) , clase abuelo
    
    #siempre defino de primero el constructor
    #En el constructor asigno los atributos del objeto que pertenece a esa clase
    
    def __init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible): #constructor
        #self.atributo=atributo
        self.capacidad_pasajeros=capacidad_pasajeros
        self.color=color
        self.tamaño=tamaño
        self.potencia=potencia
        self.combustible=combustible
        
        
    # a casi todas las clases SIEMPRE le hago 3 funciones basicas
    
    #________________________PRINT________________________________
    # IMPRIME LOS ATRIBUTOS DEL OBJETO
    
    def __str__(self):
        
        return f"atributos del objeto : {self.capacidad_pasajeros},{self.color},{ self.tamaño},{self.potencia},{self.combustible}"
       
    
    # ______________________SETTERS_______________________________
    # set -> SETUP -> CONFIGURAR O MODIFICAR ALGUN ATRIBUTO DEL OBJETO
    
    def set_capacidad_pasajeros(self,capacidad_pasajeros):
        self.capacidad_pasajeros=capacidad_pasajeros
    def set_color(self,color):
        self.color=color
    def set_tamaño(self,tamaño):
        self.tamaño=tamaño
    def set_potencia(self,potencia):
        self.potencia=potencia
    def set_combustible(self,combustible):
        self.combustible=combustible
        
    
    #________________________GETTERS______________________________
    # get -> obtener, para obtener un valor de un atributo
    
    def get_capacidad_pasajeros(self):
        return self.capacidad_pasajeros
    def get_color(self):
        return self.color
    def get_tamaño(self):
        return self.tamaño
    def get_potencia(self):
        return self.potencia
    def get_combustible(self):
        return self.combustible
    
    
#_______________subclases_________________________
 
#para heredar, pongo como argumento la clase principal (main)    
class transporte_terrestre(medios_de_transporte):
    def __init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible): #constructor
        medios_de_transporte.__init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible)
        #en al clase heredada puedo directamente llamar el constructor de la clase padre
class transporte_aereo(medios_de_transporte):
    def __init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible): #constructor
        medios_de_transporte.__init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible)
        #en al clase heredada puedo directamente llamar el constructor de la clase padre
class transporte_acuatico(medios_de_transporte):
    def __init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible): #constructor
        medios_de_transporte.__init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible)
        #en al clase heredada puedo directamente llamar el constructor de la clase padre        
        
        
#______________________________ejecucion para ver el proceso_____________________________________
carro2=transporte_terrestre(4,"blanco",8000,2000,"gasolina")   
print("carro2 :", carro2)
print(carro2.get_color())

print("conclusiones del caso1: \n 1. heredo todo \n 2. debo utilizar el constructor de la clase principal")
```

# caso 3 -> Creo nuevos atributos para la subclase y SI utilizo los atributos de la clase padre
CASO COMPLETO

```python
#_______________clase principal_________________________

class medios_de_transporte: # clase main(principal) , clase abuelo
    
    #siempre defino de primero el constructor
    #En el constructor asigno los atributos del objeto que pertenece a esa clase
    
    def __init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible): #constructor
        #self.atributo=atributo
        self.capacidad_pasajeros=capacidad_pasajeros
        self.color=color
        self.tamaño=tamaño
        self.potencia=potencia
        self.combustible=combustible
        
        
    # a casi todas las clases SIEMPRE le hago 3 funciones basicas
    
    #________________________PRINT________________________________
    # IMPRIME LOS ATRIBUTOS DEL OBJETO
    
    def __str__(self):
        
        return f"atributos del objeto : {self.capacidad_pasajeros},{self.color},{ self.tamaño},{self.potencia},{self.combustible}"
       
    
    # ______________________SETTERS_______________________________
    # set -> SETUP -> CONFIGURAR O MODIFICAR ALGUN ATRIBUTO DEL OBJETO
    
    def set_capacidad_pasajeros(self,capacidad_pasajeros):
        self.capacidad_pasajeros=capacidad_pasajeros
    def set_color(self,color):
        self.color=color
    def set_tamaño(self,tamaño):
        self.tamaño=tamaño
    def set_potencia(self,potencia):
        self.potencia=potencia
    def set_combustible(self,combustible):
        self.combustible=combustible
        
    
    #________________________GETTERS______________________________
    # get -> obtener, para obtener un valor de un atributo
    
    def get_capacidad_pasajeros(self):
        return self.capacidad_pasajeros
    def get_color(self):
        return self.color
    def get_tamaño(self):
        return self.tamaño
    def get_potencia(self):
        return self.potencia
    def get_combustible(self):
        return self.combustible
    
    
#_______________subclases_________________________
 
#para heredar, pongo como argumento la clase principal (main)    
class transporte_terrestre(medios_de_transporte):
    
    def __init__(self,n_ruedas,retrovisores,capacidad_pasajeros,color,tamaño,potencia,combustible):
        self.n_ruedas=n_ruedas
        self.retrovisores=retrovisores
        medios_de_transporte.__init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible)
        #en al clase heredada puedo directamente llamar el constructor de la clase padre
    def __str__(self):
        return f"atributos del objeto :n_ruedas={self.n_ruedas},retrovisores={self.retrovisores},capacidad_pasajeros={self.capacidad_pasajeros},color={self.color},tamaño={self.tamaño},potencia={self.potencia},combustible={self.combustible}"
        
        
    # ______________________SETTERS_______________________________
    # set -> SETUP -> CONFIGURAR O MODIFICAR ALGUN ATRIBUTO DEL OBJETO
    
    def set_n_ruedas(self,n_ruedas):
        self.n_ruedas=n_ruedas   
    def set_retrovisores(self,retrovisores):
        self.retrovisores=retrovisores      
    #________________________GETTERS______________________________
    # get -> obtener, para obtener un valor de un atributo
    
    def get_n_ruedas(self):
        return self.n_ruedas 
    def get_retrovisores(self):
        return self.retrovisores    
        
        
class transporte_aereo(medios_de_transporte):
    
    def __init__(self,altitud,n_rotores,capacidad_pasajeros,color,tamaño,potencia,combustible):
        self.altitud=altitud
        self.n_rotores=n_rotores
        medios_de_transporte.__init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible)
        #en al clase heredada puedo directamente llamar el constructor de la clase padre
    def __str__(self):
        return f"atributos del objeto :altitud={self.altitud},n_rotores={self.n_rotores},capacidad_pasajeros={self.capacidad_pasajeros},color={self.color},tamaño={self.tamaño},potencia={self.potencia},combustible={self.combustible}"
    # ______________________SETTERS_______________________________
    # set -> SETUP -> CONFIGURAR O MODIFICAR ALGUN ATRIBUTO DEL OBJETO
    
    def set_altitud(self,altitud):
        self.altitud=altitud   
    def set_n_rotores(self,n_rotores):
        self.n_rotores=n_rotores
    #________________________GETTERS______________________________
    # get -> obtener, para obtener un valor de un atributo
    
    def get_altitud(self):
        return self.altitud 
    def get_n_rotores(self):
        return self.n_rotores   
        
class transporte_acuatico(medios_de_transporte):
    
    def __init__(self,n_rotores,n_anclas,capacidad_pasajeros,color,tamaño,potencia,combustible):
        self.n_rotores=n_rotores
        self.n_anclas=n_anclas
        medios_de_transporte.__init__(self,capacidad_pasajeros,color,tamaño,potencia,combustible)
        #en al clase heredada puedo directamente llamar el constructor de la clase padre
    
    def __str__(self):
        return f"atributos del objeto :n_rotores={self.n_rotores},n_anclas={self.n_anclas},capacidad_pasajeros={self.capacidad_pasajeros},color={self.color},tamaño={self.tamaño},potencia={self.potencia},combustible={self.combustible}"
    
    # ______________________SETTERS_______________________________
    # set -> SETUP -> CONFIGURAR O MODIFICAR ALGUN ATRIBUTO DEL OBJETO
    
    def set_n_anclas(self,n_anclas):
        self.n_anclas=n_anclas   
    def set_n_rotores(self,n_rotores):
        self.n_rotores=n_rotores   
    #________________________GETTERS______________________________
    # get -> obtener, para obtener un valor de un atributo
    
    def get_n_anclas(self):
        return self.n_anclas 
    def get_n_rotores(self):
        return self.n_rotores 
    
    
#______________________________ejecucion para ver el proceso___________________________________    

#                        n_rotores,n_anclas,capacidad_pasajeros,color,tamaño,potencia,combustible
barco=transporte_acuatico(4,2,100,"rojo",800000,160000,"diesel")
print("barco1: ", barco)    


ferry=transporte_acuatico(4,2,100,"rojo",800000,160000,"diesel")
canoa=transporte_acuatico(4,2,100,"rojo",800000,160000,"diesel")
boeing777=transporte_aereo(4,2,100,"rojo",800000,160000,"diesel")
apache=transporte_aereo(4,2,100,"rojo",800000,160000,"diesel")
camioneta=transporte_terrestre(4,2,100,"rojo",800000,160000,"diesel")
motocicleta=transporte_terrestre(4,2,100,"rojo",800000,160000,"diesel")






print()    
print("conclusiones del caso3: \n 1. heredo todo, pero debo configurar el print, setters, getters especifico para cada clase, ya que tengo nuevos atributos \n 2. debo construir el constructor para cada subclase y heredar el cosntructor de la clase pirncipal")
    
```
