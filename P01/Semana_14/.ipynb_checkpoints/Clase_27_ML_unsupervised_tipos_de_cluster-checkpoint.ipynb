# Comparar algoritmos de clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.datasets import make_blobs, make_moons 
#los make dentro de sklearn.datasets son cosntructores de datos ficticios
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

# 1. Generar los datos simulados y visualizarlos

```python
help(make_blobs)
```

La función make_blobs de la biblioteca sklearn se utiliza para generar blobs gaussianos isotrópicos para la agrupación. Devuelve una matriz de muestras generadas X, una matriz de etiquetas enteras para la pertenencia a clusters de cada muestra y, y una matriz de centros de los centros de cada cluster (si return_centers=True).

El parámetro n_muestras especifica el número total de puntos a generar, que puede ser un entero (en cuyo caso los puntos se dividen por igual entre los clusters) o un array (en cuyo caso cada elemento de la secuencia indica el número de muestras por cluster).

El parámetro n_características especifica el número de características de cada muestra.

El parámetro centers especifica el número de centros a generar, o las ubicaciones fijas de los centros. Si n_muestras es un número entero y centros es Ninguno, se generan 3 centros. Si n_muestras es una matriz, centros debe ser Ninguno o una matriz de longitud igual a la longitud de n_muestras.

El parámetro cluster_std especifica la desviación estándar de los conglomerados.

El parámetro center_box especifica la caja delimitadora de cada centro de conglomerado cuando los centros se generan aleatoriamente.

El parámetro shuffle especifica si se barajan las muestras.

El parámetro random_state determina la generación de números aleatorios para la creación del conjunto de datos.

El parámetro return_centers especifica si se devuelven los centros de cada cluster.

Esta función puede utilizarse para generar conjuntos de datos sintéticos para algoritmos de agrupación, como k-means, y para probar el rendimiento de estos algoritmos.

```python
# Dataset #1.
X1, label1 = make_blobs(n_samples=200, n_features=2, centers=2, cluster_std = 5, random_state=123)
X1
```

```python
label1
```

```python
X1, label1 = make_blobs(n_samples=200, n_features=2, centers=2, cluster_std = 5, random_state=123)
#genera un conjunto de datos sintéticos con 200 puntos de datos bidimensionales, distribuidos en 2 
#cluster con una desviación estándar de 5
#X1 contiene los puntos de datos
#label1 contiene las etiquetas(target) de cluster para cada punto de datos.
plt.scatter(X1[:,0],X1[:,1], c= label1, alpha=0.7 ) # alpha es la transparencia de los puntos
#scatter : dispersion
plt.title('Dataset #1 : Original')
plt.show()
```

```python
# Dataset #2.
X2, label2 = make_moons(n_samples=200, noise=0.08, random_state=123)
plt.scatter(X2[:,0],X2[:,1], c= label2, alpha=0.7 )
plt.title('Dataset #2 : Original')
plt.show()
```

La función make_moons de sklearn se utiliza para generar un conjunto de datos sintético formado por dos semicírculos intercalados. Se suele utilizar para visualizar algoritmos de agrupación y clasificación.

Este código genera un conjunto de datos de 200 muestras con ruido añadido y traza el conjunto de datos utilizando matplotlib. El gráfico resultante muestra dos semicírculos entrelazados, con cada muestra coloreada según su etiqueta.

# 2. aplicar metodos de clustering

## Metodo k-means
La función kmeans se utiliza para agrupar datos en un número especificado de clusters utilizando el algoritmo K-means. Toma un conjunto de datos como entrada y devuelve las etiquetas de los clústeres para cada punto de datos y las coordenadas de los centros de los clústeres.

```python
# Dataset #1 and two clusters.
kmeans = KMeans(n_clusters=2,random_state=123)                 # kmeans object for 2 clusters. radom_state=123 means deterministic initialization.
kmeans.fit(X1)                                                 # Unsupervised learning => Only X1.    
myColors = {0:'red',1:'green'}                                 # Define a color palette: 0~1.
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.7 )    
plt.title('Dataset #1 : K-Means')
plt.show()
```

```python
# Dataset #2 and two clusters.
kmeans = KMeans(n_clusters=2,random_state=123)                 # kmeans object for 2 clusters. radom_state=123 means deterministic initialization.
kmeans.fit(X2)                                                 # Unsupervised learning => Only X1.    
myColors = {0:'red',1:'green'}                                 # Define a color palette: 0~1.
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.7 )
plt.title('Dataset #2 : K-Means')
plt.show()
```

## Método Aglomerative clustering

```python
# Dataset #1 and two clusters.
agglo = AgglomerativeClustering(n_clusters=2)
agglo.fit(X1)
myColors = {0:'red',1:'green'}                                 # Define a color palette: 0~1.
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(agglo.labels_).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #1 : Agglomerative')
plt.show()
```

```python
# Dataset #2 and two clusters.
agglo = AgglomerativeClustering(n_clusters=2)
agglo.fit(X2)
myColors = {0:'red',1:'green'}                                 # Define a color palette: 0~1.
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(agglo.labels_).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #2 : Agglomerative')
plt.show()
```

## Método hierarchical clustering

```python
# Dataset #1 and show dendrogram.
#linkage :vinculacion
myLinkage = linkage(X1,method='single')       # Cluster hierarchically using single linkage.
plt.figure(figsize=(20,5))
dendrogram(myLinkage)
plt.show()
```

La función de vinculación toma un conjunto de datos y un método de vinculación como entrada y devuelve una matriz de vinculación que representa la agrupación jerárquica del conjunto de datos. La función dendrograma toma la matriz de enlace y la visualiza como un diagrama de árbol, en el que cada nodo representa un conglomerado y la altura del nodo representa la distancia entre los conglomerados.

```python
from IPython.core.display import HTML
pt=pd.DataFrame(X1)
display(HTML(pt.to_html()))

```

```python
# Dataset #1 and clusters by cutting the dendrogram.
labels = fcluster(myLinkage, 5, criterion='distance')
#  Cut at the height (distance) = 5 <= change this value at will.
pd.Series(labels).value_counts()
```

La función fcluster toma como entrada una matriz de vinculación y devuelve una matriz 1D de etiquetas enteras para cada muestra de los datos originales. El segundo argumento, 5, especifica la distancia máxima entre dos muestras para que se considere que pertenecen al mismo conglomerado. El argumento criterio se establece en 'distancia', lo que significa que la distancia máxima entre dos muestras se utiliza como criterio para la agrupación.

```python
myColors = {1:'red',2:'green',3:"black",4:"blue"}                                                # Define a color palette: 1~2.
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(labels).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #1 : Hierarchical')
plt.show()
```

```python
# Dataset #2 and show dendrogram.
myLinkage = linkage(X2,method='single')       # Cluster hierarchically using single linkage.
plt.figure(figsize=(20,5))
dendrogram(myLinkage)
plt.show()
```

```python
# Dataset #2 and clusters by cutting the dendrogram.
labels = fcluster(myLinkage, 0.23, criterion='distance')                      #  Cut at the height (distance) = 0.23 <= change this value at will.
pd.Series(labels).value_counts()
```

```python
myColors = {1:'red',2:'green'}                                                # Define a color palette: 1~2.
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(labels).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #2 : Hierarchical')
plt.show()
```

## Metodo Dbscan

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es un popular algoritmo de agrupación que agrupa puntos cercanos entre sí basándose en una métrica de distancia.

El parámetro eps especifica la distancia máxima entre dos muestras para que se consideren parte del mismo cluster. El parámetro min_samples especifica el número mínimo de muestras necesarias para formar una región densa.

```python
# Dataset #1.
dbscan = DBSCAN(eps=3, min_samples=5)
dbscan.fit(X1)
myColors = {-1:'red',0:'green',1:'blue'}                                 # Define a color palette: -1~1.  Red = -1 = outlier.
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(dbscan.labels_).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #1 : DBSCAN')
plt.show()
```

```python
# Dataset #2.
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X2)
myColors = {-1:'red',0:'green',1:'blue'}                                 # Define a color palette: -1~1. Red = -1 = outlier.
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(dbscan.labels_).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #2 : DBSCAN')
plt.show()
```
