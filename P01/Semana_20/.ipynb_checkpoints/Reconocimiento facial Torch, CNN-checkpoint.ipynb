# Identificacion Facial

De forma similar a como hacemos los humanos, para que un sistema informático sea capaz de identificar a las personas que aparecen en una imagen, son necesarias varias etapas:

1. Detectar las caras en la imagen.

2. Utilizar una red neuronal capaz de mapear las características de una cara humana en una representación numérica. Este paso se conoce como embedding o encoding.

3. Medir la similitud entre la representación numérica de las caras detectadas y las representaciones de referencia disponibles en una base de datos.

4. Determinar si son suficientemente similares para considerar que pertenecen a la misma persona y asignarle la identidad.

# Detector de caras MTCNN

El primer paso en el proceso de reconocimiento facial es detectar dónde se encuentran las caras dentro de una imagen. Son muchas las estrategias y métodos de detección desarrollados desde que se iniciaran las primeras investigaciones en este campo (histogram of oriented gradients (HOG), Haar feature-based cascade, Single Shot Detector...). Uno de los últimos, y que mejores resultados consigue, es el MultiTask Cascaded Convolutional Neural Network (MTCNN). Este detector está formado por la combinación de 3 modelos de redes neuronales que van refinando, de forma secuencial, las detecciones generadas.

Existen varios detectores MTCNN para python, una de los más eficientes está implementado con pytorch y es accesible a través de la librería facenet-pytorch.

Su uso es sencillo, en primer lugar se instancia el modelo, cuyos principales parámetros son:

* image_size {int}: tamaño de la imagen de salida en pixeles (el recorte de la cara detectada). Las imágenes son cuadradas. Por defecto, el tamaño es de 160x160 pixeles.

* thresholds {list}: límites de detección para cada una de las tres redes que forman el detector MTCNN.

* margin {int}: margen de pixeles añadido la bounding box (rectángulo que rodea la cara detectada).

* min_face_size {int}: tamaño mínimo que ha de tener una cara para ser detectada.

* select_largest {bool}: si True, en el caso de que se detecten múltiples caras, se devuelve la de mayor tamaño. Si False, se devuelve la de mayor probabilidad.

* keep_all {bool}: si True, se devuelven todas las caras detectadas en la imagen, ordenadas acorde al criterio establecido por el argumento select_largest.

* post_process {bool}: si se devuelve o no la imagen normalizada.

* device {torch.device}: dónde se ejecuta el modelo. Puede ser: cpu, cuda, xpu, mkldnn, opengl, opencl, ideep, hip, msnpu, xla, vulkan.

Una vez creado el detector MTCNN, hay dos formas de aplicarlo. Con el método forward(), se devuelve un recorte de cada una de las caras detectadas. Con el método detect(), se devuelven las coordenadas de la bounding box de cada cara y, opcionalmente, los landmarks faciales(puntos clave).

```python
# Importar la biblioteca principal de PyTorch para operaciones tensoriales
import torch

# Importar la clase MTCNN de la biblioteca facenet_pytorch para la detección de rostros
from facenet_pytorch import MTCNN


# Detectar si se dispone de una GPU compatible con CUDA y configurar el dispositivo a usar
# 'cuda:0' se usa si una GPU está disponible, de lo contrario, se usa la CPU
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

'''
Una GPU CUDA (Compute Unified Device Architecture) es una unidad de procesamiento
gráfico diseñada por NVIDIA que soporta la arquitectura CUDA, un modelo de
computación paralelo desarrollado por NVIDIA. CUDA permite a los desarrolladores
utilizar el poder de procesamiento masivamente paralelo de las GPUs para tareas
de cómputo general, más allá del simple renderizado de gráficos
'''


# Imprimir el dispositivo que se está utilizando (GPU o CPU)
print('Running on device: {}'.format(device))

# Crear una instancia del detector de rostros MTCNN con parámetros específicos
mtcnn = MTCNN(
    select_largest=True,       # Seleccionar la cara más grande en la imagen (si hay varias)
    min_face_size=20,          # Tamaño mínimo de la cara a detectar (en píxeles)
    thresholds=[0.6, 0.7, 0.7],# Umbrales para las tres etapas de la red MTCNN
    post_process=False,        # Desactivar el post-procesamiento (como el recorte de bordes)
    image_size=160,            # Tamaño al que se redimensionan las imágenes de las caras detectadas
    device=device              # Dispositivo (GPU o CPU) a utilizar para la detección
)
```

# Bounding boxes y landmarks faciales


Para que el modelo MTCNN devuelva las coordenadas donde se encuentran las caras, así como los landmarks faciales, se utiliza el método detect(). Esta es la información que se necesita si lo que se pretende es dibujar un recuadro (bounding box) sobre la imagen original.

El detector MTCNN acepta como entrada imágenes PIL o numpy.ndarray 3D de tipo uint8. En el caso de disponer de múltiples imágenes, se puede pasar una lista de imágenes PIL o un batch en forma de numpy.ndarray 4D.

El resultado de la detección son 3 numpy.ndarray. Dentro de cada uno se encuentran:

* Las coordenadas de la esquina superior-izquierda y la esquina inferior-derecha de cada bounding box.

* Probabilidad de que cada bounding box contenga una cara.

* Coordenadas de los ojos, nariz y comisura de la boca en cara cada detectada.

### Una cara por imagen

```python
# Detección de bounding box y landmarks
# ==============================================================================

# Utilizar el detector MTCNN para detectar rostros en la imagen `imagen_1`
# Se detectarán las cajas delimitadoras (bounding boxes), 
# las probabilidades (confianza) y los landmarks (puntos clave del rostro)
boxes, probs, landmarks = mtcnn.detect(imagen_1, landmarks=True)

# Imprimir las cajas delimitadoras de los rostros detectados
print('Bounding boxes:', boxes)

# Imprimir las probabilidades de detección de los rostros (confianza en cada detección)
print('Probability:', probs)

# Imprimir los landmarks detectados (puntos clave del rostro como ojos, nariz y boca)
print('landmarks:', landmarks)
```

```python
# Representación con matplotlib
# ==============================================================================

# Seleccionar la primera caja delimitadora (bounding box) detectada
box = boxes[0]

# Seleccionar los landmarks (puntos clave) del primer rostro detectado
landmark = landmarks[0]

# Crear una figura y un conjunto de ejes para la visualización
fig, ax = plt.subplots(figsize=(5, 4))

# Mostrar la imagen original `imagen_1` en los ejes
ax.imshow(imagen_1)

# Dibujar los landmarks (puntos clave del rostro) en la imagen con puntos rojos
# landmark[:, 0] son las coordenadas x , landmark[:, 1] son las coordenadas y
ax.scatter(landmark[:, 0], landmark[:, 1], s=8, c='red')

# Crear un rectángulo (caja delimitadora) alrededor del rostro detectado
# El rectángulo se define por la coordenada superior izquierda (box[0], box[1])
# y las dimensiones ancho (box[2] - box[0]) y alto (box[3] - box[1])
rect = plt.Rectangle(
    xy=(box[0], box[1]),
    width=box[2] - box[0],
    height=box[3] - box[1],
    fill=False,     # No llenar el rectángulo
    color='red'     # Color del borde del rectángulo
)

# Añadir el rectángulo a los ejes para visualizar la caja delimitadora
ax.add_patch(rect)

# Ocultar los ejes para una visualización más limpia de la imagen
ax.axis('off');
```

### Múltiples caras por imágen

```python
# Detección de bounding box y landmarks
# ==============================================================================
boxes, probs, landmarks = mtcnn.detect(imagen_2, landmarks=True)

# Representación con matplotlib
# ==============================================================================
fig, ax = plt.subplots(figsize=(12, 7))
ax.imshow(imagen_2)

for box, landmark in zip(boxes, landmarks):
    ax.scatter(landmark[:, 0], landmark[:, 1], s=8, c= 'red')
    rect = plt.Rectangle(
                xy     = (box[0], box[1]),
                width  = box[2] - box[0],
                height = box[3] - box[1],
                fill   = False,
                color  = 'red'
           )
    ax.add_patch(rect)
    
ax.axis('off');
```

## Extracción de caras


Para extraer directamente las partes de la imagen que contienen caras, se utiliza el método forward(). El objeto devuelto es un tensor con el valor de los pixeles del recorte (3 canales color ° image_size ° image_size). Si se detecta más de una cara (keep_all=True), entonces se devuelve un tensor de dimensiones (nº caras x 3 canales color x image_size x image_size).

### Una cara por imagen



```python
# Detección de cara
# ==============================================================================
face = mtcnn.forward(imagen_1)
```

La imagen devuelta por el detector es un tensor con dimensiones [3, 160, 160], es decir, los canales de color están en primera posición. Para poder mostrar la imagen con matplotlib, hay que pasarlos a la última posición [160, 160, 3] y convertir de objeto tensor a numpy array.

```python
# Representación con matplotlib
# ==============================================================================
fig, ax = plt.subplots(1, 1, figsize=(3, 3))

# Cambiar el orden de las dimensiones de `face` de (C, H, W) a (H, W, C)
# C es el número de canales (por ejemplo, 3 para RGB), H es la altura y W es el ancho
# .permute(1, 2, 0) reordena las dimensiones del tensor `face`
# .int() convierte los valores del tensor a enteros
# .numpy() convierte el tensor de PyTorch a un array de NumPy
face = face.permute(1, 2, 0).int().numpy()

# Mostrar la imagen `face` en los ejes
# `face` ahora es un array de NumPy con la forma (H, W, C) adecuado para ser visualizado 
#con matplotlib
ax.imshow(face)

# Ocultar los ejes para una visualización más limpia de la imagen
plt.axis('off');
```

# Múltiples caras por imagen

```python
# Detector MTCNN
# ==============================================================================

# Crear una instancia del detector MTCNN con parámetros específicos:
# - keep_all=True: Mantener todas las detecciones de rostros, incluso si la confianza es baja
# - min_face_size=20: Tamaño mínimo de la cara a detectar en píxeles
# - thresholds=[0.6, 0.7, 0.7]: Umbrales para las tres etapas de la red MTCNN
# - post_process=False: Desactivar el post-procesamiento (como el recorte de bordes)
# - image_size=160: Tamaño al que se redimensionan las imágenes de las caras detectadas
# - device=device: Dispositivo (GPU o CPU) a utilizar para la detección
mtcnn = MTCNN(
    keep_all=True,
    min_face_size=20,
    thresholds=[0.6, 0.7, 0.7],
    post_process=False,
    image_size=160,
    device=device
)

# Detección de caras
# ==============================================================================

# Utilizar el detector MTCNN para detectar rostros en la imagen `imagen_2`
# La función forward() ejecuta la detección de caras y devuelve una lista de las caras detectadas
faces = mtcnn.forward(imagen_2)
```

Al indicar keep_all=True, se devuelven todas las caras detectadas en la imagen. Las dimensiones del tensor generado son [nº caras, canales color, altura, anchura].

```python
faces.shape
```

```python
# Representación con matplotlib
# ==============================================================================
fig, axs = plt.subplots(nrows=2, ncols=6, figsize=(10, 6))

# Aplanar la matriz de ejes en un arreglo unidimensional
axs = axs.flatten()

# Iterar sobre el número de caras detectadas en `faces`
for i in range(faces.shape[0]):
    # Obtener la i-ésima cara de la matriz de caras detectadas `faces`
    face = faces[i, :, :, :].permute(1, 2, 0).int().numpy()
    
    # Mostrar la imagen de la cara en el i-ésimo eje
    axs[i].imshow(face)
    
    # Ocultar los ejes para una visualización más limpia de la imagen
    axs[i].axis('off')

# Ajustar automáticamente la disposición de los ejes para evitar superposiciones
plt.tight_layout()
```

El proceso de extracción también puede hacerse utilizando las coordenadas de las bounding box. La diferencia respecto a extraerlas directamente con el detector MTCNN es que, este último, hace un redimensionamiento para que la imagen devuelta sea cuadrada.

# Embedding de caras


Una vez identificadas las caras de la imagen, el siguiente paso consiste en obtener una transformación numérica que sea capaz de representar los aspectos característicos de cada una. Al vector numérico resultante se le conoce como embedding o encoding.

Los modelos de deep learning (redes convolucionales) que permiten obtener el embedding de caras no son sencillos de entrenar. Afortunadamente, existen varios modelos ya entrenados que son accesibles desde python. Dos de ellos son:

* face_recognition_model_v1 de la librería dlib. Este modelo genera un embedding de 128 valores.
    * https://github.com/davisking/dlib/blob/master/python_examples/face_recognition.py
* facenet_pytorch import InceptionResnetV1 de la librería facenet_pytorch. Este modelo genera un embedding de 512 valores.
    * https://github.com/timesler/facenet-pytorch
    
Para obtener este tipo de modelos, primero se entrena un modelo de clasificación con múltiples personas. Una vez entrenada la red, se elimina la última capa (softmax) para que la salida del modelo sea un vector numérico. En este documento se utiliza el modelo InceptionResnetV1, en concreto, la versión entrenada con el set de datos vggface2. Para conocer más detalles sobre este tipo de modelos, es recomendable consultar la publicación VGGFace2.
https://arxiv.org/pdf/1710.08092.pdf


```python
# Modelo para hacer el embedding de las caras
# ==============================================================================

# Importar la clase InceptionResnetV1 del módulo facenet_pytorch
from facenet_pytorch import InceptionResnetV1
'''
# Crear una instancia del modelo InceptionResnetV1 para generar embeddings de caras
# - pretrained='vggface2': Utilizar los pesos preentrenados en el conjunto de datos VGGFace2
# - classify=False: Desactivar la clasificación (solo se extraen embeddings,
    no se realiza clasificación)
# - device=device: Especificar el dispositivo (GPU o CPU) en el que se ejecutará el modelo
# - .eval(): Poner el modelo en modo de evaluación para desactivar la regularización dropout
    y batch normalization

'''

encoder = InceptionResnetV1(pretrained='vggface2', classify=False, device=device).eval()
```

```python
# Detección de cara
# ==============================================================================
cara = mtcnn(imagen_1)
```

```python
# Embedding de cara
# ==============================================================================
embedding_cara = encoder.forward(cara.reshape((1,3, 160, 160))).detach().cpu()
embedding_cara
```

# Similitud entre caras


El objetivo de obtener una representación numérica de las caras (embeddings) es poder cuantificar similitudes entre ellas. Dos formas de calcular esta similitud es utilizando la distancia euclídea o la distancia coseno entre embeddings. Cuanto menor es la distancia, mayor la similitud de las caras.

Para el ejemplo se compara una imagen frente a otras dos, la primera pertenece a la misma persona, Phil Dunphy, y la segunda a Cameron Tucker.

```python
# Detector MTCNN
# ==============================================================================

# Crear una instancia del detector MTCNN con parámetros específicos:
# - keep_all=True: Mantener todas las detecciones de rostros, incluso si la confianza es baja
# - min_face_size=20: Tamaño mínimo de la cara a detectar en píxeles
# - thresholds=[0.6, 0.7, 0.7]: Umbrales para las tres etapas de la red MTCNN
# - post_process=False: Desactivar el post-procesamiento (como el recorte de bordes)
# - image_size=160: Tamaño al que se redimensionan las imágenes de las caras detectadas
# - device=device: Dispositivo (GPU o CPU) a utilizar para la detección
mtcnn = MTCNN(
    keep_all=True,
    min_face_size=20,
    thresholds=[0.6, 0.7, 0.7],
    post_process=False,
    image_size=160,
    device=device
)
```

```python
# Extracción de las caras utilizando MTCNN
# ==============================================================================

# Utilizar el detector MTCNN para detectar y recortar la cara en la imagen `imagen_1`
# Se toma la primera cara detectada (índice 0) y se almacena en la variable phil_1
phil_1 = mtcnn(imagen_1)[0]

# Utilizar el detector MTCNN para detectar y recortar la cara en la imagen `imagen_2`
# Se toma la tercera cara detectada (índice 1) y se almacena en la variable phil_2
phil_2 = mtcnn(imagen_2)[1]

# Utilizar el detector MTCNN para detectar y recortar la cara en la imagen `imagen_2`
# Se toma la segunda cara detectada (índice 2) y se almacena en la variable cameron
cameron = mtcnn(imagen_2)[2]
```

```python
# Representación con matplotlib
# ==============================================================================
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(10, 6))

face = phil_1.permute(1, 2, 0).int().numpy()
axs[0].imshow(face)
axs[0].set_title('Phil 1')
axs[0].axis('off')

face = phil_2.permute(1, 2, 0).int().numpy()
axs[1].imshow(face)
axs[1].set_title('Phil 2')
axs[1].axis('off')

face = cameron.permute(1, 2, 0).int().numpy()
axs[2].imshow(face)
axs[2].set_title('Cameron')
axs[2].axis('off');
```

Una vez extraídas las 3 caras de las imágenes, se crean sus embeddings y se calcula la distancia euclídea entre ellas.

```python
# Embeddings
# ==============================================================================
'''
# Generar el embedding para la cara de "phil_1" utilizando el modelo "encoder" 
    (InceptionResnetV1)
# Se redimensiona la cara a un tensor de tamaño (1, 3, 160, 160) para que coincida 
    con el formato esperado por el modelo
# .reshape((1, 3, 160, 160)) cambia la forma de la cara a un tensor con dimensión (1, 3, 160, 160)
# .detach() se utiliza para evitar que se calculen gradientes durante la propagación hacia atrás
# .cpu() mueve el tensor al CPU (si se estaba utilizando la GPU)

'''

embeding_phil_1 = encoder.forward(phil_1.reshape((1, 3, 160, 160))).detach().cpu()

# Generar el embedding para la cara de "phil_2" utilizando el modelo "encoder" (InceptionResnetV1)
# Se sigue el mismo proceso que para "phil_1"
embeding_phil_2 = encoder.forward(phil_2.reshape((1, 3, 160, 160))).detach().cpu()

# Generar el embedding para la cara de "cameron" utilizando el modelo "encoder" (InceptionResnetV1)
# Se sigue el mismo proceso que para "phil_1"
embeding_cameron = encoder.forward(cameron.reshape((1, 3, 160, 160))).detach().cpu()
```

```python
# Distancias entre embeddings de caras
# ==============================================================================
# Convertir los embeddings en vectores unidimensionales
embeding_phil_1_flat = embeding_phil_1.flatten()
embeding_phil_2_flat = embeding_phil_2.flatten()
embeding_cameron_flat = embeding_cameron.flatten()



# Importar la función 'euclidean' de la biblioteca scipy.spatial.distance
from scipy.spatial.distance import euclidean

# Distancia entre cada imagen
distance_same_phil = euclidean(embeding_phil_1_flat, embeding_phil_1_flat)
distance_two_phil = euclidean(embeding_phil_1_flat, embeding_phil_2_flat)
distance_phil_cameron = euclidean(embeding_phil_1_flat, embeding_cameron_flat)


# Imprimir la distancia euclidiana entre el mismo embedding de Phil 
# (distancia entre un punto y él mismo)
print(f"Distancia entre la misma imagen Phil: {distance_same_phil:.10f}")
print(f"Distancia entre las dos imágenes de Phil: {distance_two_phil:.10f}")
print(f"Distancia entre Phil y Cameron: {distance_phil_cameron:.10f}")
```

Para poder identificar a quién pertenece una cara, es necesario contrastarla frente a una base de datos que contenga un embedding de referencia por cada identidad. Si no se dispone del vector de referencia de una identidad, esta no puede ser identificada.
